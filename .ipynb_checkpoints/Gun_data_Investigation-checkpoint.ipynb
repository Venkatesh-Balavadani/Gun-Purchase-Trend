```python
#importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import calendar
%matplotlib inline
```

```python
#reading the csv files
gun_dataFrame = pd.read_csv('gun_data.csv')
```

```python
census_dataFrame = pd.read_csv('U.S. Census Data.csv')
```

```python
#getting the required data from the dataset
census_dataFrame = census_dataFrame.iloc[1:4]
```

```python
#renaming the index
census_dataFrame.reset_index(drop=True,inplace=True)
```

```python
#renaming the records in fact column
census_dataFrame.insert(loc=1,column='year',value=['estimate_base','population_2016','population_2010'])
```

```python
census_dataFrame.drop(labels='Fact',axis=1,inplace=True)
```

```python
#taking transpose to match the gun dataset
census_dataFrame = census_dataFrame.T
```

```python
census_dataFrame.head(3)
```

```python
census_dataFrame.columns=['estimate_base','population percent','population 2010']
```

```python
census_dataFrame.drop(labels='year',axis=0,inplace=True)
```

```python
census_dataFrame['population 2010'] = census_dataFrame['population 2010'].astype(str)
census_dataFrame['population 2010']
```

```python
census_dataFrame['population 2010'] = census_dataFrame['population 2010'].str.replace(",","")
```

```python
census_dataFrame['population 2010'] = census_dataFrame['population 2010'].astype(float)
```

```python
census_dataFrame['population percent'] = census_dataFrame['population percent'].astype(float)
```

```python
census_dataFrame['population percent'] = census_dataFrame['population percent'].apply(lambda x:x+1)
```

```python
census_dataFrame['estimate_base'] = census_dataFrame['estimate_base'].astype(str)
```

```python
census_dataFrame['estimate_base'] = census_dataFrame['estimate_base'].str.replace(",","")
```

```python
census_dataFrame['estimate_base'] = census_dataFrame['estimate_base'].astype(float)
```

```python
population_2016 = census_dataFrame['estimate_base']*census_dataFrame['population percent']
```

```python
census_dataFrame['population_2016'] = population_2016.astype(int)
```

```python
#starting gun dataset
month_value = gun_dataFrame['month'].str.split("-").str[1]
year_value = gun_dataFrame['month'].str.split("-").str[0]
```

```python
gun_dataFrame.insert(loc=1,column='year',value=year_value)
```

```python
gun_dataFrame.insert(loc=2,column='months',value=month_value)
```

```python
gun_dataFrame.head()
```

```python
gun_dataFrame['months'] = gun_dataFrame['months'].astype(int)
```

```python
gun_dataFrame['months'] = gun_dataFrame['months'].apply(lambda x: calendar.month_abbr[x])
```
